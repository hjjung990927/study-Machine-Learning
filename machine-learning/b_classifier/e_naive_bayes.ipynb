```python
import pandas as pd

spam_df = pd.read_csv('./datasets/spam.csv')
spam_df
```

```python
spam_df.info()
```

```python
spam_df.isna().sum()
```

```python
# ì¤‘ë³µ ì œê±°
spam_df.duplicated().sum()
```

```python
spam_df.drop_duplicates(inplace=True)
```

```python
spam_df = spam_df.reset_index(drop=True)
spam_df
```

```python
from sklearn.preprocessing import LabelEncoder

spam_encoder = LabelEncoder()
targets = spam_encoder.fit_transform(spam_df.Category)
spam_df['Target'] = targets
spam_df
```

```python
spam_df.drop(labels=['Category'], axis=1, inplace=True)
spam_df
```

```python
spam_df.Target.value_counts()
```

```python
targets_0 = spam_df[spam_df.Target == 0].sample(641, random_state=124)
targets_1 = spam_df[spam_df.Target == 1]

spam_df = pd.concat([targets_0, targets_1])
spam_df
```

```python
spam_df.reset_index(drop=True, inplace=True)
spam_df.Target.value_counts()
```

```python
spam_df
```

#### CountVectorizer
- ë¬¸ì¥ì— ìˆëŠ” ë‹¨ì–´ë“¤ì— ì¸ë±ìŠ¤ë¥¼ ë¶™ì—¬ì„œ ê° ë‹¨ì–´ì˜ ë¹ˆë„ìˆ˜ë¥¼ ì„¸ì–´ì£¼ëŠ” ê¸°ìˆ 

```python
from sklearn.feature_extraction.text import CountVectorizer

text = [
    "í–„ë²„ê±°ëŠ” ë§›ìˆì–´! ì •ë§ ë§›ìˆì–´",
    "ì•„ë‹ˆì•¼ í”¼ìê°€ ë” ë§›ìˆì–´ í–„ë²„ê±°ë³´ë‹¤ ë” ë§›ìˆì–´!",
    "ì•„ë‹ˆì•¼ ë‘˜ ë‹¤ ë¨¹ì!"    
]

c_vct = CountVectorizer()
freq = c_vct.fit_transform(text)
print(freq.toarray())
print(c_vct.vocabulary_)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
train_test_split(spam_df.Message, spam_df.Target, stratify=spam_df.Target, test_size=0.2)
```

```python
y_test.value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

m_nb_pipe = Pipeline(
    [
        ('count_vectorizer', CountVectorizer()), ('multinomail_NB', MultinomialNB())
    ])
m_nb_pipe.fit(X_train.values, y_train)
```

```python
# prediction = m_nb_pipe.predict(X_test.values)
m_nb_pipe.score(X_test.values, y_test)
```

```python
print(m_nb_pipe.predict(['Free drink. Contact us']))
print(m_nb_pipe.predict(["Please let me know if you need any additional information regarding the project."]))
print(m_nb_pipe.predict(["Weâ€™re excited to offer you an exclusive discount on our new product line this season."]))
```
