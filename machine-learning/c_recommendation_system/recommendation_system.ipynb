```python
import numpy as np

s1 = np.array([1, 0, 1])
s2 = np.array([1, 1, 1])
```

```python
from numpy import dot
from numpy.linalg import norm

def get_cosine_simliarity(A, B):
    return dot(A, B) / (norm(A) * norm(B))

get_cosine_simliarity(s1, s2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from kiwipiepy import Kiwi
sentences = [
    "나는 밥을 먹었다",
    "나는 빵을 먹었다",
    "나는 밥과 빵을 먹는다"
]

kiwi = Kiwi()
josa_tags = ["JKS","JKC","JKO","JKG","JKB","JKV","JKQ","JX","JC",
            "EC","EF","EP","ETN","ETM"]

def removes_josa(sentence):
    result = []
    for token in kiwi.tokenize(sentence):
        if token.tag in josa_tags:
            continue

        result.append(token.lemma)
    return " ".join(result)

sentences = list(map(removes_josa, sentences))

cvt = CountVectorizer()
vectors = cvt.fit_transform(sentences).toarray()

cosine_similarity(vectors)
```

```python
import pandas as pd
import numpy as np

m_df = pd.read_csv('./datasets/movie_dataset.csv')
m_df.info()
```

```python
columns = ['keywords', 'overview', 'tagline', 'cast', 'director']
pre_m_df = m_df[columns].fillna("")
pre_m_df.isna().sum()
```

```python
columns = ['keywords', 'overview', 'tagline', 'cast', 'director']
features = pre_m_df.loc[:, columns]

result_df = features.keywords + \
features.overview + \
features.tagline + \
features.cast + \
features.director
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_v = CountVectorizer()
count_matrix = count_v.fit_transform(result_df)

cosine = cosine_similarity(count_matrix)
print(cosine.shape)
```

```python
m_df
```

```python
# 사용자가 입력한 영화, 전체 영화 목록
movie_user_likes = "Men in Black"
movie_index = m_df[m_df.title == movie_user_likes].index[0]
similar_movies = list(enumerate(cosine[movie_index]))
sorted_similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)[1:]

print(f'{movie_user_likes}를 보셨다면 추천하는 영화 3편\n')
for item in sorted_similar_movies[:5]:
    print(m_df[m_df.index == item[0]]['title'].values[0])
    # print(m_df[m_df.index == item[0]]['title'])
```
